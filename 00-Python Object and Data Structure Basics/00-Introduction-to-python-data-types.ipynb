#  Data Types

| Name          | Type | Description |
|--------------|------|-------------|
| Integers     | `int` | Whole numbers without a decimal point, e.g. `3`, `300`, `200` |
| Floating Point | `float` | Numbers containing a decimal point, e.g. `2.3`, `4.6`, `100.0` |
| Strings      | `str` | Ordered sequence of characters enclosed in quotes, e.g. `"hello"`, `"Sammy"`, `"2000"`, `"楽しい"` |
| Lists        | `list` | Ordered, mutable collection of objects; can hold mixed data types, e.g. `[10, "hello", 200.3]` |
| Dictionaries | `dict` | Unordered collection of key–value pairs used for fast lookup, e.g. `{"mykey": "value", "name": "Frankie"}` |
| Tuples       | `tuple` | Ordered, immutable collection of objects; similar to lists but cannot be changed, e.g. `(10, "hello", 200.3)` |
| Sets         | `set` | Unordered collection of **unique** objects; duplicates are not allowed, e.g. `{"a", "b"}` |
| Booleans     | `bool` | Logical values representing truth conditions: `True` or `False` |

## Types of numbers

Python has various "types" of numbers (numeric literals). We'll mainly focus on integers and floating point numbers.

Integers are just whole numbers, positive or negative. For example: 2 and -2 are examples of integers.

Floating point numbers in Python are notable because they have a decimal point in them, or use an exponential (e) to define the number. For example 2.0 and -2.1 are examples of floating point numbers. 4E2 (4 times 10 to the power of 2) is also an example of a floating point number in Python.

### Basic Arithmetic

```python
# Addition
2+1
```

```python
# Subtraction
2-1
```

```python
# Multiplication
2*2
```

```python
# Division
3/2
```

```python
# Floor Division
7//4
```

The reason we get this result is because we are using "*floor*" division. The // operator (two forward slashes) truncates the decimal without rounding, and returns an integer result.

## Modulo or "Mod" Operator (%)

```python
#Modulo
3 % 2
```

 The % operator returns the remainder after division.

Use case: If we have to check the either number is even or odd, we use mod operator i.e., if remainder is 0 it is even and if remainder is not 0 then its odd.

```python
# Powers
2**3
```

```python
# Can also do roots this way
4**0.5
```

```python
# Order of Operations followed in Python
2 + 10 * 10 + 3
```

```python
# Can use parentheses to specify orders
(2+10) * (10+3)
```
